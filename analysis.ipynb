```python
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
zodiac_counts = {}
total_file_count = 0
for filename in glob.iglob('blogs/' + '**', recursive=True):
    if filename.endswith(".xml"):
        total_file_count += 1
        blogger_attributes = filename.split('.')
        if zodiac_counts.get(blogger_attributes[-2], False):
            zodiac_counts[blogger_attributes[-2]] += 1
        else:
            zodiac_counts[blogger_attributes[-2]] = 1
```

```python
zodiac_counts
```

```python
sum(zodiac_counts.values())
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

cmap = plt.cm.tab10
colors = cmap(np.arange(len(zodiac_counts)) % cmap.N)

ax.bar(zodiac_counts.keys(), zodiac_counts.values(), color=colors)

ax.set_xlabel('Zodiac Signs')
ax.set_ylabel('Number of Bloggers')

plt.show()
```
